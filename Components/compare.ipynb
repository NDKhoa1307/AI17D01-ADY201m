```python
import requests
import json 
import pandas as pd
import pymongo
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Calling api and converting the data into readable form

url = f'https://archive-api.open-meteo.com/v1/archive?latitude=16.0678&longitude=108.2208&start_date=1993-09-30&end_date=2023-09-30&daily=weathercode,temperature_2m_mean,precipitation_sum,precipitation_hours,windspeed_10m_max,windgusts_10m_max,et0_fao_evapotranspiration&timezone=auto'
response = requests.get(url)
data = json.loads(response.text)
```

```python
# Use json_normalize to turn data into pandas dataframe

df = pd.json_normalize(data)

# Use pandas methods to transform the dataframe

df = df.apply(pd.Series.explode)
df = df.reset_index(drop = True)
df.head()
```

```python
# Modify columns name

df.columns = df.columns.str.replace('daily.', 'data-')
df.columns = df.columns.str.replace('daily_units.', 'unit-')
df.head()
```

```python
df.columns
```

```python
# If want to save to MongoDB database then change this value into True

installedMongo = False
if installedMongo:
    # Connect to localhost database
    client = pymongo.MongoClient("mongodb://localhost:27017")
    db = client['weather']
    collection = db['weather-daily']
    
    # Convert dataframe
    insertData = df.to_dict(orient="records")   
    
    # Insert into MongoDB collection
    inserted_ids = collection.insert_many(insertData).inserted_ids

    client.close()
else:
    if not os.path.isdir('../data'):
        os.makedirs('../data')

    df.to_csv('../data/dailyWeather.csv', index=False)
```

```python
df['data-time'] = pd.to_datetime(df['data-time'])


df['year']=df['data-time'].dt.year

yearly_rainfall=df.groupby(['year'])['data-precipitation_sum'].sum()
yearly_hours_rain=df.groupby(['year'])['data-precipitation_hours'].sum()
yearly_eva=df.groupby(['year'])['data-et0_fao_evapotranspiration'].sum() 
yearly_temp=df.groupby(['year'])['data-temperature_2m_mean'].mean()
```

```python
# Plotting rainfall and Precipitation Hours
fig, ax1 = plt.subplots(figsize=(12, 6))
# Plot Yearly Rainfall, Precipitation Hours as bars
ax1.bar(yearly_rainfall.index - 0.2, yearly_rainfall, width=0.2, label='Yearly Rainfall')
ax1.bar(yearly_hours_rain.index, yearly_hours_rain, width=0.2, label='Yearly Precipitation Hours')
# Set axis labels and title
ax1.set_xlabel('Year')
ax1.set_ylabel('Rainfall, Precipitation Hours')
ax1.set_title('Yearly Climate Trends')
ax1.legend(loc='upper left')


# Plotting Evapotranspiration and temp 
fig, ax3 = plt.subplots(figsize=(12, 6))
# Plot Yearly Evapotranspiration as bars
ax3.bar(yearly_eva.index + 0.2, yearly_eva, width=0.2, label='Yearly Evapotranspiration')
# Set axis labels and title
ax3.set_xlabel('Year')
ax3.set_ylabel(' Evapotranspiration')
ax3.set_title('Yearly Climate Trends')
# Create a second y-axis for temperature
ax4 = ax3.twinx()
ax4.plot(yearly_temp.index, yearly_temp, color='red', label='Yearly Mean Temperature')
ax4.set_ylabel('Mean Temperature', color='brown')
ax4.tick_params(axis='y', labelcolor='brown')
ax3.legend(loc='upper left')
ax4.legend(loc='upper right')


# Show the plot
plt.show()




```
